```python
# Excelファイルの読み込み
file_path = "第11屆平地原住民立法委員選舉候選人在臺北市各投開票所得票數一覽表.xlsx"
excel_file = pd.ExcelFile(file_path)

print(excel_file.sheet_names)
sheet_name = excel_file.sheet_names[0]
df = excel_file.parse(sheet_name)

df.iloc[1] = df.iloc[1].fillna(df.iloc[0])
df_cleaned = df.dropna(how='all')

# タイトル行の削除（2行目がデータの始まりと仮定）
df_cleaned = df_cleaned.iloc[1:]  # 1行目（インデックス0）を削除

# カラム名をデータの最初の行にリセット
df_cleaned.columns = df_cleaned.iloc[0]  # 現在の最初の行をカラム名に設定
df_cleaned = df_cleaned.iloc[1:]  # カラム名として使った行を削除

# 結果を確認
df_cleaned.head()
```

```python
df_cleaned = df.drop(columns=[col for col in df.columns if "nan" in col])

# Attempt to convert all columns except "鄉(鎮、市、區)別" and "村里別" to numeric types
for col in df_cleaned.columns:
    if col not in ["鄉(鎮、市、區)別", "村里別"]:
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# Display the cleaned DataFrame and check for remaining non-numeric columns
df_cleaned.info()
df_cleaned.head()
```
